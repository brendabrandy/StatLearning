# Introduction

Cybersecurity has become a more pressing issues in the past decade. It is a 1.3 billion dollar industry, and its target customers ranges from the daily internet user to the military base. There has also been research gearing towards providing better security for individuals and organizations. To facilitate such research, MIT Lincoln Laboratory and DARPA jointly developed a set of data in 1998 to simulate an attacked network of the US Air Force One Research lab. 

In the simulated network, 2 subnetworks - an external and an internal network - are connected through a CISCO router. The external network contains all the websites of the internet, as well as an attacker that tries to infiltrate the lab through the network, while the internal network computers with 3 victim OS - Solaris, Linux and Windows.

The simulation lasted for a total of eight weeks. The first six weeks are used as training data, while the last two weeks are used as test data. There is a total of four million data points and I am going to perform some exploratory data analytics on them

# Notebook and Data Prep

First we import the libraries necessary. Pandas are used to provide a schema for the data, while seaborn is used to visualize data

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import kendalltau
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from matplotlib import gridspec
import pylab
```

Now let's import the data and name each column with the provided schema. 

```python
# Setting up the dataframe
attack_df = pd.read_csv("KDDTrain+_20Percent.txt", sep=",",header = None)
attack_df.columns = ["duration","protocol_type","service","flag","src_bytes","dst_bytes","land",
                     "wrong_fragment","urgent","hot","num_failed_logins","logged_in","num_compromised",
                     "root_shell","su_attempted","num_root","num_file_creations","num_shells",
                     "num_access_files","num_outbound_cmds","is_host_login","is_guest_login","count",
                     "srv_count","serror_rate","srv_serror_rate","rerror_rate","srv_rerror_rate",
                     "same_srv_rate","diff_srv_rate","srv_diff_host_rate","dst_host_count",
                     "dst_host_srv_count","dst_host_same_srv_rate","dst_host_diff_srv_rate",
                     "dst_host_same_src_port_rate","dst_host_srv_diff_port_rate","dst_host_serror_rate",
                     "dst_host_srv_serror_rate","dst_host_rerror_rate","dst_host_srv_rerror_rate",
                     "attack_type","score"]

# From inspections of the data frame, the attack_types accidentally included a fullstop at the end
# Hence these fullstops needs to be removed

attack_df["attack_type"] = attack_df["attack_type"].str.strip(".")
print attack_df.attack_type.unique()
```

There are a total of 23 kinds of attack -- including normal connections. To make data processing easier, the KDD Cup further generalize the attacks -- leading down to only 4 attacks:

    - User to Root (U2R) : Changing operation mode from user mode to root mode
    - Denial of Service (DoS):
    - Remote to Local (R2L):
    - Probe : Scanning the network to obtain information, usually for surveillance purposes

Hence we can map each attack to the 4 general attacks

```python
gen_attack_dict= {'buffer_overflow':'u2r', 
                  'loadmodule':'r2l', 
                  'perl':'u2r', 
                  'neptune':'dos',
                  'smurf':'dos', 
                  'guess_passwd':'r2l', 
                  'pod':'dos', 
                  'teardrop':'dos', 
                  'portsweep':'probe', 
                  'ipsweep':'probe',
                  'land':'dos', 
                  'ftp_write':'r2l', 
                  'back':'dos', 
                  'imap':'r2l', 
                  'satan':'probe', 
                  'phf':'r2l', 
                  'nmap':'probe',
                  'multihop':'r2l', 
                  'warezmaster':'r2l', 
                  'warezclient':'r2l', 
                  'spy':'r2l', 
                  'rootkit':'u2r',
                  'normal':'normal'}
attack_df['gen_type'] = attack_df['attack_type'] .map(gen_attack_dict)
attack_df.head(10)
```

# Distribution of Attacks

Visualizing the distribution of attacks can give us a better impression of the data distribution. This is important because we need to know how many data points we would have for each attack. 

```python
attack_series = attack_df['gen_type'].value_counts()
attack_gen_pie = attack_series.plot(kind='pie',figsize=(5,5),
                 title='Distribution of general types of attacks')
```

From the distribution, we can see that there is an even distribution between normal and abnormal connections, however, amongst the abonormal connections, the sample set is very imbalanced. Let's look closer at all the abnormal connections. 

```python
ano_attack_df = attack_df[attack_df['attack_type'] != "normal"]
ano_attack_series = ano_attack_df['attack_type'].value_counts()
ano_attack_gen_pie = ano_attack_series.plot(kind='pie', figsize=(5,5),
                     title="Distribution of Malicious connections")
```

From the data, now we can see why DoS occupies such a large percentage of connections when compared to the rest. Looking at the largest portion, neptune, it essentially flood a computer with a lot of SYN signals until the stack on the server that handles SYN commands overflows and goes down. The mechanism of the attack itself utilizes a lot of connections to successfully launch the attack. 

# Network Protocol of Attacks

The first thing we should understand about attacks is what channel does the attack communicate through

```python
print attack_df.protocol_type.unique()
```

The three unique channels are tcp, udp and icmp. It would be interesting to see what attacks go through certain channels and how often do said attacks do so. 

```python
ano_attack_df = attack_df[attack_df.attack_type!="normal"]
gen_protocol = ano_attack_df.groupby(["gen_type","protocol_type"])
gen_protocol_freq = gen_protocol.size().reset_index().rename(columns={0:'freq'})
g_gen_protocol = sns.barplot(x='protocol_type',y='freq',hue='gen_type',data=gen_protocol_freq)
g_gen_protocol_final = g_gen_protocol.set(title="Occurences of General Exploits in Various Protocols",
                xlabel="Protocol",
                ylabel="Frequency (log N)",
                yscale="log")
```

The above graph is a general impression of attacks and protocols. It is interesting to see the only attacks that uses the TCP protocol is R2L attacks, and that U2R attacks utilizes the ICMP protocol. If you look very closely, you would see a small amount of u2r attacks utilizes the udp protocol. The following graph shows a more detailed description of each protocol vs each attack

```python
protocol = ano_attack_df.groupby(["attack_type","protocol_type"])
protocol_freq = protocol.size().reset_index().rename(columns={0:'freq'})
g_protocol = sns.barplot(x='protocol_type',y='freq',
                         hue='attack_type',data=protocol_freq)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
g_protocol_final = g_protocol.set(title="Occurences of Exploits in Various Protocols",
                xlabel="Protocol",
                ylabel="Frequency (log N)",
                yscale="log")
```

Personally, I found the low usage of udp very interesting. it seems that the only attacks that use udp are smurf, multihop and teardrop. UDP (User Datagram Protocol) is an alternative communications protocol to Transmission Control Protocol (TCP) used primarily for establishing low-latency and loss tolerating connections between applications on the Internet. There must be a reason that these attacks utilize UDP, primarily because of the advantages that UDP offers.

# Level of Successful Prediction

When the data is generated, the research group also implemented their own learners to see whether they can learn from the provided data. There's a total of 21 learners, and they documented the number of successfully learned attacks. It would be interesting to see how successful the predictions are for each attack.

```python
df_slim_data = attack_df[[
                        "gen_type",
                       "score"]]
# df_slim_abnormal = df_slim_data[df_slim_data["gen_type"] != "normal"]
# data = pd.melt(df_slim_abnormal, 
""""num_root","num_file_creations","num_shells",
                     "num_access_files","num_outbound_cmds","""
               # id_vars=["gen_type"], 
               # var_name="stats")
data = df_slim_data
data.head(10)
```

```python
norm = data[data["gen_type"] == "normal"]
dos = data[data["gen_type"] == "dos"]
probe = data[data["gen_type"] == "probe"]
r2l = data[data["gen_type"] == "r2l"]
u2r= data[data["gen_type"] == "u2r"]
plt.hist(norm["score"],alpha=0.5, label="normal")
plt.hist(dos["score"],alpha=0.5, label="dos")
plt.hist(probe["score"],alpha=0.5, label="probe")
plt.hist(r2l["score"],alpha=0.5, label="r2l")
plt.hist(u2r["score"],alpha=0.5, label="u2r")
plt.legend(loc="upper left")
```

The ones that have the highest successful prediction rates are normal and DoS connections. For probe attacks, the attack is less successfully detected. Zooming in closer...

```python
plt.hist(r2l["score"],alpha=0.5, label="r2l")
plt.hist(u2r["score"],alpha=0.5, label="u2r")
plt.legend(loc="upper left")
```

U2R is hardly identified, while there's something like a normal distribution for the identification of R2L. It doesn't surprise me that R2L and U2R are hardly identified because there's less data about them, while there's ample information about normal and DoS connections.
